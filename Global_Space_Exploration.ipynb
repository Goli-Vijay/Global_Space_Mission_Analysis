## Exploratory Data Analysis

### Number of Missions by Country

```python
top_countries = df['Country'].value_counts().head(10)
print(top_countries)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index,hue=top_countries.index,palette='viridis')
plt.title("Top 10 Countries by Number of Space Missions")
plt.xlabel("Number of Missions")
plt.ylabel("Country")
plt.show()
```

### Number of Missions Over Time (Yearly Trend)

```python
missions_per_year = df['Year'].value_counts().sort_index()
print(missions_per_year)

plt.figure(figsize=(10, 6))
sns.lineplot(x=missions_per_year.index, y=missions_per_year.values, marker="o")
plt.title("Number of Space Missions Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Missions")
plt.grid(True)
plt.show()
```

### Mission Types Distribution

```python
plt.figure(figsize=(8,5))
df['Mission Type'].value_counts().plot(kind='pie', autopct='%1.2f%%',startangle = 40,shadow=True)
plt.title("Mission Type Distribution")
plt.show()
```

### Budget distribution

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Budget(in billion $)'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Budget (in billion $)")
plt.xlabel("Budget (in billion $)")
plt.ylabel("Number of Missions")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Budget Analysis by Country

```python
budget_by_country = df.groupby('Country')['Budget(in billion $)'].sum().sort_values(ascending=False).head(10)
print(budget_by_country)

plt.figure(figsize=(12, 6))
sns.barplot(x=budget_by_country.values, y=budget_by_country.index,hue = budget_by_country.index ,palette='coolwarm')
plt.title("Top 10 Countries by Total Mission Budget")
plt.xlabel("Total Budget (in Billion $)")
plt.ylabel("Country")
plt.show()
```

### Success Rate Analysis by Country

```python
success_rate_country = df.groupby('Country')['Success Rate'].mean().sort_values(ascending=False).head(10)
print(success_rate_country)

plt.figure(figsize=(12, 6))
sns.barplot(x=success_rate_country.values, y=success_rate_country.index,hue = success_rate_country ,palette='cool')
plt.title("Top 10 Countries by Average Success Rate")
plt.xlabel("Average Success Rate (%)")
plt.ylabel("Country")
plt.show()
```

### Environmental Impact Distribution

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Environmental Impact', order=df['Environmental Impact'].value_counts().index,hue = 'Mission Type' , palette='Blues')
plt.title("Environmental Impact Distribution")
plt.xlabel("Impact Level")
plt.ylabel("Number of Missions")
plt.show()
```

```python
tech_counts = df['Technology Used'].value_counts()
print(tech_counts)
```

### Technology Usage Distribution

```python
plt.figure(figsize=(5,5))
plt.pie(tech_counts,labels=tech_counts.index,autopct='%1.1f%%',startangle=140)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title("Technology Usage Distribution")
plt.axis('equal')
plt.show()
```

## Bivariate Analysis

### Budget vs Country  Top 5 Spending Countries

```python
top_budget = df.groupby('Country')['Budget(in billion $)'].sum().sort_values(ascending=False).head(5)
budget_df = top_budget.reset_index()
print(budget_df)

sns.barplot(data=budget_df,x='Country',y='Budget(in billion $)',hue = 'Budget(in billion $)',legend = False,palette='dark')
plt.title("Top 5 Countries by Total Budget")
plt.xlabel("Budget (in Billion $)")
plt.ylabel("Country")
plt.show()
```

### Mission Type vs Environmental Impact

```python
grouped = df.groupby(['Mission Type', 'Environmental Impact']).size().reset_index(name='Count')
print(grouped)
```

```python
sns.barplot(data=grouped,x='Environmental Impact',y='Count',hue='Mission Type',palette='dark')
plt.title("Environmental Impact vs Mission Type")
plt.xlabel("Environmental Impact")
plt.ylabel("Number of Missions")
plt.legend(title='Mission Type')
plt.show()
```

### Technology Used vs Success Rate

```python
box_data = df.groupby('Technology Used')['Success Rate'].describe()
print(box_data)
```

```python
sns.boxplot(data=df,x='Technology Used',y='Success Rate',hue = 'Technology Used',legend=False,width=0.6,palette='Set1')
plt.title("Success Rate by Technology Used")
plt.xticks(rotation=45)
plt.show()
```

### Satellite Type vs Duration

```python
sat_duration = df.groupby('Satellite Type')['Duration'].agg(['mean', 'max'])
print(sat_duration)
```

```python
sns.violinplot(data=df,x='Satellite Type',y='Duration',hue = 'Satellite Type',palette='dark')
plt.title("Duration by Satellite Type")
plt.xticks(rotation=45)
plt.show()
```

## Multivariate Analysis

```python
sns.pairplot(df[['Budget(in billion $)', 'Success Rate', 'Duration']])
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```

```python
corr = df[['Budget(in billion $)', 'Success Rate', 'Duration']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Mission Type', y='Success Rate', hue='Technology Used')
plt.title("Success Rate by Mission Type and Technology Used")
plt.xticks(rotation=45)
plt.legend(title='Technology Used')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Satellite Type', y='Duration', hue='Environmental Impact')
plt.title("Duration by Satellite Type & Environmental Impact")
plt.xticks(rotation=45)
plt.show()
```
